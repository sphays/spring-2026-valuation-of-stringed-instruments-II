# EDA with adjusted prices
We will perform data analysis after sale prices have been adjusted for inflation. After gathering insights, we then hope to run a random forest regression to determine feature importance.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn

from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import TargetEncoder, OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import root_mean_squared_error


```

We will drop columns we do not plan to use, and create a decade and weekday column to see if these features are sensitive to sale prices. We also decide to look at the prices after a log transformation.

```python
df = pd.read_csv(r"C:\Users\Sean\Documents\GitHub\spring-2026-valuation-of-stringed-instruments-II\price_adjusted_data.csv")
df = df.drop(columns = ['Unnamed: 0', 'usd', 'gbp', 'eur', 'cpi_usd', 'cpi_eur', 'gbp_usd', 'eur_usd', 'cpi_gbp', 'maker_name' ] )
df.columns = df.columns.str.lower()
df['decade'] = (df['year']//10) * 10

df['weekday'] = pd.to_datetime(df['sale_date']).dt.day_name()
df['price_usd_real'] = np.log(df['price_usd_real'])
df.head()
```

Next, we find the mean prices for various feature values and plot their bar graphs

```python
meanprices_day = df.groupby(df['weekday'])['price_usd_real'].mean()
meanprices_decade = df.groupby(df['decade'])['price_usd_real'].mean()
meanprices_month = df.groupby(df['month'])['price_usd_real'].mean()
meanprices_house = df.groupby(df['auction_house'])[['price_usd_real']].mean()

fig, axes = plt.subplots(nrows=2, ncols =2, figsize = (10,4) )
meanprices_day.plot(ax=axes[0,0], kind='barh')
meanprices_decade.plot(ax=axes[0,1], kind = 'barh')
meanprices_house.plot(ax=axes[1,0], kind = 'bar')
meanprices_month.plot(ax=axes[1,1], kind = 'bar')
plt.tight_layout
plt.show()
```

It appears that auction house plays an important role in price. Prices in the earliest two decades are significantly larger than other decades; we can examine this more by looking at price by year. Weekday and month seem less indicative of price, so we will try to look at mean prices by day instead.

```python
meanprices_year = df.groupby(df['year'])['price_usd_real'].mean()
meanprices_day = df.groupby(df['day'])['price_usd_real'].mean()

fig, axes = plt.subplots(nrows=1, ncols =2, figsize = (10,7) )

meanprices_year.plot(ax=axes[0], kind='barh')
meanprices_day.plot(ax=axes[1], kind = 'barh')
plt.show()
```

By year, there is a substantial drop in prices after 1982. Prices by day seem almost uniform. Next, we will take random samples of instrument makers and plot their sale prices over time. We well also look at instrument prices by type with respect to year.

```python
# selecting makers with over 200 sales for a more dense graph
df['maker_freq'] = df.groupby(df['maker_id'])['maker_id'].transform('count')
makers_filter = df[df['maker_freq'] >= 200]

random_makers = df['maker_id'].sample(n=10).tolist()
random_makers_df = df.loc[df['maker_id'].isin(random_makers) ]
SalesOverTime = random_makers_df.pivot_table(index = 'sale_date', columns = 'maker_id', values = 'price_usd_real' )
ax = SalesOverTime.plot()
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout
plt.show()

TypeSalesTime = df.pivot_table(index = 'year', columns = 'type', values = 'price_usd_real')
ax = TypeSalesTime.plot()
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

For some samples of maker id's, there does not seem to be a major correlation between sale date and prices. It is possible a maker's individual instruments have a better correlation, but we do not currently have data on instrument id's. Some instrument types seem to exhibit a sale price increase over time.

```python
# you can check the linear correlation for each sample
random_makers_df['year'].corr(df['price_usd_real'] )
```

We can place makers and cities into bins based on their frequency and visualize their mean prices.

```python
df['city_freq'] = df.groupby('city_maker')['city_maker'].transform('count')
df['maker_freq'] = (df['maker_freq']//50) * 50
df['city_freq'] = (df['city_freq']//100) * 100
meanprices_maker = df.groupby('maker_freq')['price_usd_real'].mean()
meanprices_city = df.groupby('city_freq')['price_usd_real'].mean()

fig, axes = plt.subplots(nrows=1, ncols =2, figsize = (8,5) )

meanprices_maker.plot(ax=axes[0], kind='barh')
meanprices_city.plot(ax=axes[1], kind = 'bar')
plt.show()
```

The frequency that a city shows up in our dataframe appears to give some information of the mean price. It is possible that "bold_currency", which describes the original currency a sale was made in, could be relevant.

```python
meanprice_bold = df.groupby('bold_currency')['price_usd_real'].mean()
print(meanprice_bold)
```

Now we will attempt to build a random forest regression model with a selection of the previous features, with the aim of seeing which ones were most important in the model evaluation. We will use target encoding for high cardinality features and hot one encoding for low cardinality features.

```python
df = df.drop(columns = ['month', 'maker_freq', 'day'] )
df.head()
```

```python
x = df.iloc[:, df.columns != 'price_usd_real']
y = df.iloc[:, df.columns == 'price_usd_real']

x_train , x_test, y_train , y_test = train_test_split(x,y, test_size = 0.2, random_state = 42)
```

```python
tec_features = ['maker_id', 'lot', 'auction_house', 'sale_date', 'city_maker', 'decade', 'type', 'year' ]
hoc_features = ['bold_currency', 'weekday']
preprocessor = ColumnTransformer(
    transformers = [
        ('tec_features', TargetEncoder(), tec_features ),
        ('hoc_features', OneHotEncoder(), hoc_features  )
    ],
    remainder = 'passthrough'
)

pipeline = Pipeline(steps = [
    ('preprocessor', preprocessor),
    ('classifier',  RandomForestRegressor(n_estimators=300, random_state=42, oob_score=True))
])
pipe = pipeline.fit(x_train,y_train)

preds = pipe.predict(x_test)
```

```python

rmse = root_mean_squared_error(y_test, preds)
print(rmse)
```

Because the model has performed poorly, it is not valuable to investigate which features the model used to make its decisions. After experimenting, it seemed that including 'sale_date' and 'bold_currency" improved performance. Although, not by much. A few missing variables affecting prices might be the physical condition of the instruments, a unique instrument identifier, and the age of each instrument.
